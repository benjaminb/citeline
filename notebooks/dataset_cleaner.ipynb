```python
import pandas as pd
import json

with open("../data/datasets/acl200_global/processed/acl_titlemap.json", "r") as f:
    title_map = json.load(f)
full_data = pd.read_csv("../data/datasets/acl200_global/context_dataset_modified.csv")
```

```python
# Confirm that every citing title and every target title is in the title map
for i, row in full_data.iterrows():
    has_citing_title = row['citing_title'] in title_map
    has_target_title = row['target_title'] in title_map
    miss_str = ""
    if not has_citing_title:
        miss_str += "citing "
    if not has_target_title:
        miss_str += "target "
    if miss_str:
    # if not has_citing_title or not has_target_title:
        print(f"{i}: missing: {miss_str.strip()}")

print("Done checking titles.")
```

```python
from acl_anthology import Anthology

anthology = Anthology.from_repo() 

title_to_ref = {str(p.title): p for p in anthology.papers()}
for ds_title, title in title_map.items():
    if title not in title_to_ref:
        print(f"Title not found in anthology: {title}")
        print(f"Dataset title: {ds_title}")
        print("---")
print("Done mapping titles to references.")
```

```python
# Every title maps to a paper reference with a full id and pdf url
for title in set(title_map.values()):
    paper_ref = title_to_ref.get(title)
    paper_id = paper_ref.full_id
    paper_pdf_url = getattr(paper_ref, "pdf", None)
    message = ""
    if not paper_id:
        message += " no id "
    if not paper_pdf_url:
        message += " no pdf "
    if message:
        print(f"Title: {title} has issues:{message}")
print("Done checking paper references.")
```
