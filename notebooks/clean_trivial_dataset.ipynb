```python
import json

infile_path = '../data/dataset/trivial_llm.jsonl'
outfile_path = '../data/dataset/trivial_cleaned.jsonl'
```

```python
with open(infile_path, 'r') as infile, open(outfile_path, 'w') as outfile:
    for line in infile:
        example = json.loads(line)
        if example['sent_no_cit'].strip() == example['sent_original'].strip():
            json.dump(example, outfile)
            outfile.write('\n')
```
