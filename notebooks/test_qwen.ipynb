```python
from citeline.embedders import Embedder
```

```python
embedder_06 = Embedder.create(model_name="Qwen/Qwen3-Embedding-0.6B", device="mps", normalize=True)
```

```python
embeddings = embedder_06(
    [
        "what are the best rides at disney world",
        "people love the teacup ride at disney world",
        "be sure to wear suntan lotion at the beach",
    ]
)
```

```python
embeddings[0] @ embeddings[1].T
```

```python
embeddings[0] @ embeddings[2].T
```
