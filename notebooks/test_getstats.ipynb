```python
import glob

directory = "experiments/query_expansions/results_qwen8/chunks"
paths = glob.glob(
    "../" + directory + "/**/results_*.json", recursive=True)

print(f"{'Experiment':35s} | Hitrate@25   | Recall@25    | Hitrate@50   | Recall@50")
for p in paths:
    with open(p, "r") as f:
        results = json.load(f)
    label = results["config"].get("plot_label", "no label")
    hitrate_at_25 = results["average_hitrate_at_k"][24]
    recall_at_25 = results["average_recall_at_k"][24]
    hitrate_at_50 = results["average_hitrate_at_k"][49]
    recall_at_50 = results["average_recall_at_k"][49]
    print(f"{label:35s} | H@25: {hitrate_at_25:.4f} | R@25: {recall_at_25:.4f} | H@50: {hitrate_at_50:.4f} | R@50: {recall_at_50:.4f}")
```
