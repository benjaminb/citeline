This takes the existing search results (a 21G file) which was created with the qwen06/add02/interleave search strategy before the metrics and rank fusion code was refactored, and adds the required fields.

```python
import json
from tqdm import tqdm
```

```python
input_path = "../experiments/multiple_query_expansion/results/search_results.jsonl"
output_path = "../data/search_results_add02_interleaved.jsonl"

with open(input_path, 'r') as infile, open(output_path, 'w') as outfile:
    for line in tqdm(infile):
        data = json.loads(line)
        record = data['record']
        record['query'] = record.pop('sent_no_cit')
        record['query_add_prev_2'] = record.pop('add_prev_2')
        data['record'] = record

        outfile.write(json.dumps(data) + '\n')
```
