```python
import arxiv

search = arxiv.Search(query="ti:\"attention is all you need\"", max_results=3)
```

```python
client_results = arxiv.Client().results(search)
top_results = list(client_results)
for result in top_results:
    print(result.title)
    print(result.authors)
    print(result.published)
    print(result.pdf_url)
    print("=====")
```

```python
paper = top_results[0]
print(paper.get_short_id())
```

```python
record = {
    "title": paper.title,
    "doc_id": paper.get_short_id(),
    "pubdate": paper.published,
    "url": paper.pdf_url,
    "url_2": paper.download_source,
    "url_3": paper.source_url,
    "other_links": paper.links
}

for key, value in record.items():
    print(f"{key}: {value}")
```
