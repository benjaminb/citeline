```python
# import json
from pprint import pprint
from semantic_text_splitter import TextSplitter
from utils import load_dataset

data = load_dataset('data/json/Astro_Reviews.json')
print(len(data))
```

```python
splitter = TextSplitter((200, 1500)) # chunks will be in this range
chunks = splitter.chunks(data[0]['abstract'])
print(chunks)
```

```python
pprint(chunks)
```

```python
chunks = splitter.chunks(data[0]['body'])
pprint(chunks)
```

```python
default_splitter = TextSplitter(1500, overlap=150)
body_chunks_default = default_splitter.chunks(data[0]['body'])
pprint(body_chunks_default)
```
