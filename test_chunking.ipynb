```python
# import json
from pprint import pprint
from semantic_text_splitter import TextSplitter
from utils import load_dataset

data = load_dataset('data/json/Astro_Reviews.json')
```

```python
def make_full_text(record):
    return record['title'] + '\n\nABSTRACT:\n' + record['abstract'] + '\n\n' + record['body']
```

```python
default_splitter = TextSplitter(1500, overlap=150)
chunks = default_splitter.chunks(make_full_text(data[0]))
print(len(chunks))
for chunk in chunks:
    print(len(chunk))
```
