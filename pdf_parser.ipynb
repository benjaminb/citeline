```python
import PyPDF2
```

```python
texts = []
with open('data/pdf/BERT.pdf', 'rb') as file:
    reader = PyPDF2.PdfReader(file)


    for page_num in range(len(reader.pages)):
        page = reader.pages[page_num]
        text = page.extract_text()
        texts.append(text)

full_text = '\n'.join(texts)
```

```python
from pprint import pprint
pprint(full_text[:1306])
```

```python
import transformers

pipeline = transformers.pipeline(
    "text-generation",
    model="microsoft/phi-4",
    model_kwargs={"torch_dtype": "auto"},
    device_map="auto",
)

messages = [
    {"role": "system", "content": "You are a medieval knight and must provide explanations to modern people."},
    {"role": "user", "content": "How should I explain the Internet?"},
]

outputs = pipeline(messages, max_new_tokens=128)
print(outputs[0]["generated_text"][-1])
```

```python
messages_bert = [
    {"role": "system", "content": "You are an expert researcher and proofreader. You will read the following text and answer the question: what new contributions does this paper make?"},
    {"role": "user", "content": full_text[:1306]},
]

outputs_bert = pipeline(messages_bert, max_new_tokens=128)
pprint(outputs_bert[0]["generated_text"][-1])
```

```python
import transformers
import torch

model_id = "meta-llama/Llama-3.1-8B"

pipeline = transformers.pipeline(
    "text-generation", model=model_id, model_kwargs={"torch_dtype": torch.bfloat16}, device_map="auto"
)

pipeline("Hey how are you doing today?")
```

```python
result = pipeline('List the original contributions described in this research paper abstract:\n' + full_text[:1306], 
                  max_length=4096)
```
